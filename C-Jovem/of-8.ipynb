```python
# Importações 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Gerando as notas
def geraNota(semente, turma):
    np.random.seed(semente)
    alunos = [f"T{turma} - Aluno {i+1}" for i in range(30)]
    notas = np.random.randint(0, 11, size=30)
    return alunos, notas
```

```python
# Gerando o data frame turma 1
alunos, notas = geraNota(41, 1)
df_turma1 = pd.DataFrame({
    "Aluno - T1": alunos,
    "Nota - T1": notas
})
df_turma1.head()
```

```python
# Gerando o data frame turma 2
alunos, notas = geraNota(42, 2)
df_turma2 = pd.DataFrame({
    "Aluno - T2": alunos,
    "Nota - T2": notas
})
df_turma2.head()
```

```python
# Gerando as notas com letras
def geraNotaLetras(semente):
    np.random.seed(semente)
    alunos = [f"T3 - Aluno {i+1}" for i in range(30)]
    notas_letras = np.random.choice(['A', 'B', 'C'], size=30)
    return alunos, notas_letras
```

```python
# gerando o dataframe
alunos, notas_letras = geraNotaLetras(42)
df_turma3 = pd.DataFrame({
    "Aluno - T3": alunos,
    "Nota - T3": notas_letras
})
df_turma3
```

```python
# Padroniza de acordo com o as notas numéricas 
df_turma3.loc[df_turma3['Nota - T3'] == 'A', 'Nota - T3'] = 10
df_turma3.loc[df_turma3['Nota - T3'] == 'B', 'Nota - T3'] = 8
df_turma3.loc[df_turma3['Nota - T3'] == 'C', 'Nota - T3'] = 6

df_turma3.head()
```

```python
# concatenado os data frames
df = pd.concat([df_turma1, df_turma2, df_turma3], axis=1)
df
```

```python
# Calculando as médias das turmas
media_turma1 = df["Nota - T1"].mean()
media_turma2 = df["Nota - T2"].mean()
media_turma3 = df["Nota - T3"].mean()
```

```python
# Criando um DataFrame para as médias
desempenho = pd.DataFrame({
    "Turma": ["Turma 1", "Turma 2", "Turma 3"],
    "Média": [media_turma1, media_turma2, media_turma3]
})
```

```python
# Plotando o gráfico de barras
plt.bar(desempenho["Turma"], desempenho["Média"], color=["blue", "green", "red"])
plt.title("Desempenho Geral das Turmas")
plt.ylabel("Média das Notas")
plt.xlabel("Turmas")
plt.ylim(0, 10)
plt.show()
```
