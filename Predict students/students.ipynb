# **Preveja a evasão e o sucesso acadêmico dos alunos**
Investigando o Impacto de Fatores Sociais e Econômicos

## Sobre este conjunto de dados
Este conjunto de dados fornece uma visão abrangente dos alunos matriculados em vários cursos de graduação oferecidos em uma instituição de ensino superior. Inclui dados demográficos, fatores socioeconômicos e informações de desempenho acadêmico que podem ser usados ​​para analisar os possíveis preditores de evasão e sucesso acadêmico. Este conjunto de dados contém vários bancos de dados disjuntos que consistem em informações relevantes disponíveis no momento da inscrição, como modo de inscrição, estado civil, curso escolhido e muito mais. Adicionalmente, estes dados podem ser utilizados para estimar o desempenho global dos alunos no final de cada semestre, através da avaliação das unidades curriculares creditadas/inscritas/avaliadas/aprovadas, bem como das respetivas classificações. Finalmente, temos a taxa de desemprego, a taxa de inflação e o PIB da região, o que pode nos ajudar a entender melhor como os fatores econômicos influenciam as taxas de evasão estudantil ou os resultados de sucesso acadêmico. Esta poderosa ferramenta de análise fornecerá informações valiosas sobre o que motiva os alunos a permanecer na escola ou abandonar seus estudos para uma ampla gama de disciplinas, como agronomia, design, educação, enfermagem, jornalismo, gerenciamento, serviço social ou tecnologias

```python
# Cria um dataframe com os valores null
def columnsNull(dataset):
    df_nulos = pd.DataFrame(dataset.isnull().sum())
    valor = []
    indice = []
    for i, r in df_nulos.iterrows():
        if r[0] > 0:
            indice.append(i)
            valor.append(r[0])
    df_nulos = pd.DataFrame(np.array(valor).reshape(1, -1), columns=indice)
    return df_nulos

columnsNull(ds)
```

### Dividido a base de treino

#### Bibliotecas

```python
# Libs
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# Classes e previsores 
X = df.iloc[:, 0:13].values
y = df.iloc[:, 13].values
y
```

```python
# Divisão da base de dados entre treinamento e teste 
# (30% para testar e 70% para treinar)
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(
    X, y, test_size=0.3, random_state=42
)
X_treinamento.shape, X_teste.shape
```

### Modelo de Machine Learning

```python
# Regressão Logística
rl = LogisticRegression(solver='lbfgs', max_iter=1000)
rl.fit(X_treinamento, y_treinamento)
rl_pred = rl.predict(X_teste)
print(f'Taxa de acertos: {accuracy_score(y_teste, rl_pred)}');
```

```python
# Matriz de confusão 
cm = confusion_matrix(y_teste, rl_pred)
cm
```
